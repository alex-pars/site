```python
import numpy as np
import math as mat


# Stochastic catolgue  
# Input lambda and Total number of years  lambda min = 0.02
# Randomly generate time intervals from exponential distribution with lambda_min

# Calculate time of first event
i = 0
lam = 0.02  # Annual rate of eq occurrence
Totyear = 10000   # Set no. of years to produced stochastic catalogue for  

Time_interval = []
val = (-mat.log(1-np.random.uniform(low=0.0, high=1.0, size=None)))/lam    # Pr(X=<x) = 1-exp(-lam*x)
Time_interval.append(val)

#print(Time_interval)

Event_time = []
Event_time.append(val)



while Event_time[i] < Totyear: # Calculate event intervas using exponential distribution: stop calculating events when total event time is larger than the catalogue range
    i = i + 1
    Time_interval.append((-mat.log(1-np.random.uniform(low=0.0, high=1.0, size=None)))/lam)
    Event_time.append(Event_time[i-1] + Time_interval[i])

#print(Event_time)

No_events = len(Event_time) 

print(No_events,"events in simulated catalogue.")

# Generic check = N_event = Nyear * lambda 
CHECK = Totyear * lam 
#print(CHECK)

```

```python

# Plot of simulation histogram and analytical exponential distribution


x = np.arange(0, 400, 1)

Pr_cata = 1 - np.exp(-lam * x)


# dPr_cata= diff(Pr_cata)   %could just say dPr_cata= lam*exp(-lam*x)
dPr_cata = lam * np.exp(-lam * x) 

import matplotlib.pyplot as plt

fig1 = plt.figure()
plt.hist(Time_interval, bins = 25, density = True, color = "skyblue", label='Simulation histogram')
plt.plot(x,dPr_cata, 'r', label='Exponential distribution')
plt.legend(frameon = False)
plt.xlabel('Time interval X_t (years)') 
plt.ylabel('PDF')
plt.show()
```
